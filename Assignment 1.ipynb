```python
# 10 Combination of Data Operators
```

```python
a = [9,17,22]
b = [3,9,6]
print("a+b=",(a[0]+b[0]))
print("a-b=",(a[1]-b[2]))
print("a*b=",(a[2]*b[1]))
print("a/b=",(a[0]/b[0]))
print("a//b=",(a[1]//b[2]))
print("a%b=",(a[2]%b[1]))
print("a>b=",(a[0]>b[0]))
print("a<b=",(a[1]<b[2]))
print("a==b=",(a[2]==b[1]))
print("a!=b=",(a[0]!=b[0]))
```
